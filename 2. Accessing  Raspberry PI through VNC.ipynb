# Accessing Raspberry PI through VNC

First install `tightvncserver`

```python
!sudo apt-get install tightvncserver
```

Then start the VNC Server at 720p

```python
!tightvncserver -geometry 1280x720
```

Create a password to use the VNC session, set this to be the same as your pi login. There is no need to set a view only password. The Raspberry Pi has now started a VNC server session at `raspberrypi.local:1` remember this address as it is needed later.

Install the VNC Viewer from the official site https://www.realvnc.com/en/connect/download/viewer/%20

Then accept the warnning and give the password configured before.



## Editing corntab to start the VNC when reboot


First enter the next command in the terminal

```python
!sudo crontab -e
```

Edit the crontab adding the following line of code at the end. `@reboot su - pi -c '/usr/bin/tightvncserver -geometry 1280x720'`

Press CTRL + X, Y then Enter to save and exit. Now when the Pi is booted up, the VNC server will start.
