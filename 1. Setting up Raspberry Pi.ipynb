# Setting up Raspberry PI

## Installing the Image of the Raspberry PI and configuring it to be accessible throught SSH

We need first to install the image of the Raspberry PI on the SD card. For that, we are going to donwload the Raspberry PI Imager. https://www.raspberrypi.com/software/

We can also set some Advance opntions up with the command CTRL+SHIFT+X. Here we can enable the SSH, creating a user called `pi` and set a password for this user. We can also connect the rapberry to the internet network

Then before turning the Rapsberry Pi on. We need to insert the SD card and the camera on the raspberry pi. The following image shows and example of how to connect the camera.

![Alt Text](https://projects-static.raspberrypi.org/projects/getting-started-with-picamera/380e9259139db923be4170f89920ff1975a2357c/en/images/connect-camera.gif)

## Accessing the Raspberry Pi

To get the address of the Raspberry PI. We can only do a simple ping to the address `raspberrypi.local`

In this case the ip adress of the Raspberry Pi is going to be `192.168.0.604` and it can easily be accessed by using the command `ssh <user_name>@<ip_address>`

```python
!ssh pi@192.168.0.204
```

Then is going to ask us for the password we set before during the installation of the image on the SD card. Once set, we have access to the Raspberry Pi
