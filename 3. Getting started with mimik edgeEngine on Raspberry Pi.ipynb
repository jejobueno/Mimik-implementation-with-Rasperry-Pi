# Getting started with Mimik edgeEngine within Raspberry Pi

First of all, our raspberry should have installed `npm`, `node`, `wget` and `git`. Normally `wget` and `git` are already pre-installed with the OS. Let's isntall `node` and `npm`

````python
!sudo apt install npm
````

Once complete, we verify the versions of npm and node:

````python
!npm version
````

## Openning port `8083`

The edgeEngine Microservices listen by default to port 8083, for open this port we can do it using the program `Unconplicated Firewal [uwf]`

````python
!sudo apt install ufw
````

We check our version:

````python
!sudo apt --ufw version
````

Then we allow the port `8083` on the Rasberry device using the following command:

````python
!sudo ufw allow 8083
````

We sould get as response:
```
   Rules updated
   Rules updated (v6)
```

## Donwloading the latest release od edgeEngine Runtime for Rapsbian

We need to get the files from the latest git repository to start the edgeEngine.

````python
!wget https://github.com/edgeEngine/egdeengine-raspbian/releases/download/v2.3.1/edgeEngine-raspbian-setup-v2.3.1.tar
````

Then extract the files from  the tar using

````python
!tar xvf edgeEngine-raspbian-setup-v2.3.1.tar
````

And now start the edgeEgnine excecuting the `./start.sh` script

````python
setsid ./start.sh
````

## Installing the edge-engine-cli tool

First donwload and install the mimik-edge-engine tool using the following command:

````python
!sudo npm install -g @mimik/mimik-edge-cli
````

Check that the tool has been correctly installed using the next command

````python
!mimik-edge-cli -v
````

You will get a response like the following:
```
    @mimik/mimik-edge-cli 1.1.0 linux-x64 node-v14.17.5
```

Execute the following command to test that the edgeEngine is accessible trough the mimik-edge-cli:

````python
!mimik-edge-cli account get-me
````

You should get the following response or similar
